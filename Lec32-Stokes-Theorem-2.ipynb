# Stokes Theorem

A curve C bounds a surface S and orient the curve and surface compatible with each other. Then we can compute line integral along C

$$\oint_c \vec F \cdot d\vec r = \iint_S (curl \vec F)\cdot \hat n\ dS$$

### Stokes and Path independence

Defined:

A region is simply connected if every closed loop inside it bounds a surface inside it

Example: A region of entire space with origin removed is simply connected. We can find any surface bounded by a closed loop. If that surface passes through origin, just push it a little bit so that it does not pass through origin

Example: Space with z-axis removed is not simply connected. A unit circle in the xy plane. If we try to find a surface whose boundary is that circle, it has to cross z-axis somewhere. For a donut shape, it wouldn't be a surface with boundary just that circle

Recall: if $\vec F = \nabla F$ is a gradient, then $curl \vec F = 0$. The mixed second partial derivatives are equal. 

Theorem: if $\vec F$ is defined in a simply connected region and $curl \vec F = 0$, then $\vec F$ is a gradient field and $\int_c \vec F\cdot d\vec r$ is path independent

Proof: Assume $curl \vec F = 0$, show 

$$\int_{c1} \vec F \cdot d\vec r - \int_{c2} \vec F \cdot d\vec r = 0$$

Let c = c1 + c2

$$\int_{c1} \vec F \cdot d\vec r - \int_{c2} \vec F \cdot d\vec r = \oint_c \vec F \cdot d\vec r$$

We can find S because the region is simply connected

Using Stokes

$$= \iint_S (curl \vec F)\cdot d\vec S = \iint_S 0 \cdot d\vec S = 0$$

## Stokes and Surface Independence

A closed curve C. We can take bound surface S1 from top like a down with normal vector points outwards or from bottom like a bowl with normal vector points inwards

Stokes says $$\oint_c \vec F \cdot d\vec r = \iint_{s1} (\nabla \times \vec F)\cdot \hat n\ dS = \iint_{s2 }(\nabla \times \vec F)\cdot \hat n\ dS$$

This equality seems curl has some sort of surface independence property. It doesn't care about the bounded surface, as long as its boundary is curve C

Why same?

$$\iint_{s1} (\nabla \times \vec F)\cdot \hat n\ dS - \iint_{s2} = \iint_{s=s1-s2} (\nabla \times \vec F)\cdot \hat n\ dS $$

By divergence theorem (S closed surface)

$$= \iiint_D div(\nabla \times \vec F)\ dV = 0$$

Can check: $div(\nabla \times \vec F) = 0$ always. So flux for s1 and s2 is the same

$$\iint_{s1} = \iint_{s2}$$

If the vector field $\vec F = \langle P,Q,R \rangle$, the curl is defined by the cross product 

$$\nabla \times \vec F = \begin{vmatrix}\hat i & \hat j & \hat k \\
\frac{\partial }{\partial x} & \frac{\partial }{\partial y} & \frac{\partial }{\partial z} \\
P & Q & R
\end{vmatrix} = \langle R_y-Q_z, P_z-R_x, Q_x-P_y \rangle$$

$$div(\nabla \times \vec F ) = (R_y-Q_z)_x + (P_z-R_x)_y + (Q_x-P_y)_z$$

$$= R_{yx} - Q_{zx} + P_{zy} - R_{xy} + Q_{xz} - P_{yz} = 0$$ 

So, the divergence of a curl is zero

$$\nabla \cdot (\nabla \times \vec F)  = 0$$

Note: for real vectors $u \cdot (u \times v) = 0$. Cross product gives normal direction, dot product with 90 degree vector becomes zero


## Review

### Triple Integral

$$\iiint_R f\ dV$$

- rectangular: $dV = dz\ dx\ dy$

- cylindrical: $dV = dz\ r\ dr\ d\theta$

The main trick is to find the bounds of integration. If do it with dz first, for a fixed point in xy plane, look at the bound for z. Figure out the bottom surface and top surface of the solid

After that, find bounds for x and y. Look at its projection on the plane, set up double integral either in rectangular or cylindrical coordinate 

- spherical: $dV = \rho^2 \sin\phi\ d\rho\ d\phi\ d\theta$

Applications:

- Mass
- Average value of f
- Moment of Inertia
- Gravitational attraction on mass at origin


### Double Integral

$$\iint_S \vec F \cdot \hat n\ dS$$

Formula for $\hat n\ dS$

Ex: $\hat n\ dS = ... \ dx\ dy$ the double integral becomes $\iint ...\ dx\ dy$

- horizontal plane inside yz plane. 

$dS = dy\ dz$

$\hat n = \pm\hat i$

- sphere centered at origin

$$\hat n = \pm \frac{\langle x,y,z \rangle}{radius}$$

$$dS = a\ dz\ d\theta$$

- cylinders centered on z-axis

$$\hat n = \pm \frac{\langle x,y,0 \rangle}{radius}$$

$$dS = a^2 \sin\phi\ d\phi\ d\theta$$

General case:

if we know $z = z(x,y)$, then $\hat n\ dS = \pm \langle -z_x, -z_y, 1 \rangle \ dx\ dy$

if $\vec N$ given normal vector: $\hat n\ dS = \pm \frac{\vec N}{\vec N \cdot \hat k}\ dx\ dy$


### Line Integral

$$\int_c \vec F \cdot d\vec r = \int_c P\ dx + Q\ dy + R\ dz$$

parameterize c -> express in terms of a single variable


Different evaluation techniques for triple, double and line integral. There are connections between them. Triple and double integral is connected by divergence theorem. Double and line integral is connected by Stokes theorem


### Divergence Theorem

Divergence theorem says if there is a region D in space, bounded by boundary S. So, it is a closed surface, orient with normal vector n pointing outwards, then surface integral can be replaced with triple intetgral

$$\iint_s \vec F \cdot \hat n\ dS = \iiint_D (div \vec F)\ dV$$

LHS is a vector field, and RHS is a function that somehow are related. We will be able to compute both side separately


### Stokes Theorem

If there is a closed curve C that is a boundary of surface S and orient with normal vector n in a compatible manner , then line integral of C can be replaced by surface integral on S

$$\oint_C \vec F \cdot d\vec r = \iint_S (\nabla \times \vec F)\cdot \hat n\ dS$$

This relates line integral of one field to another field in surface integral. The second field (on RHS) is obtained by taking curl of the first field (on LHS). They way to compute RHS is just like the way to compute any surface integral. Find value of $\hat n\ dS$ dot product and evaluate


If you have a function in its gradient, fundamental theorem of calculus says that the line integral for the vector field given by the gradient of a function is equal to the value change of a function. 

$$\int_C \nabla f \cdot d\vec r = f(p_0) - f(p_1)$$

p0 and p1 are the start and end point of the curve C

These three theorems relate a quantity with a certain number of integral signs to a quantity with one more integral signs
