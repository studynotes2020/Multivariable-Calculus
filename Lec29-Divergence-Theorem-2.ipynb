# Divergence Theorem

$$\unicode{x222F}_S \vec F \cdot d\vec S = \iiint_D div \vec F\ dV$$

To evaluate using coordinate system:

$$\unicode{x222F}_S \langle P, Q, R \rangle \cdot \hat n\ dS = \iiint_D (P_x + Q_y + R_z)\ dV$$


### $\nabla$ notation

$\nabla$ read del

$$\nabla = \langle \frac{\partial}{\partial x}, \frac{\partial}{\partial y}, \frac{\partial}{\partial z} \rangle$$

Gradient

$$\nabla f = \langle \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y}, \frac{\partial f}{\partial z} \rangle$$

Function goes in, vector goes out

Divergence

$$\nabla \cdot \vec F = \langle \frac{\partial}{\partial x}, \frac{\partial}{\partial y}, \frac{\partial}{\partial z} \rangle \cdot \langle P, Q, R \rangle = \langle \frac{\partial P}{\partial x}, \frac{\partial Q}{\partial y}, \frac{\partial R}{\partial z} \rangle$$

Vector goes in, function goes out

## Physical Interpretation of Divergence

$div(\vec F) =$ source rate = amount of flux generated per unit volume

Incompressible fluid flow: given mass occupies a fixed volume. We have such a fluid with velocity $\vec F$

$\iiint_D div \vec F \ dV = \unicode{x222F}_S \vec F \cdot \hat n\ dS = $ amount of fluid leaving D per unit time


## Proof of Divergence Theorem

$$\unicode{x222F}_S \langle 0,0,R \rangle \cdot \hat n \ dS = \iiint_D R_z\ dV$$

then get the general case by summing three such identities for each component

If region D is vertically simple where it lives between top $z = z_2(x,y)$ and bottom $z = z_1(x,y)$

U is the area D projected on xy plane

Right hand side: $$\iiint_D R_z\ dV = \iint_U \int_{z_1}^{z_2} R_z\ dz\ dx\ dy = \iint_U [R(x,y,z_2) - R(x,y,z_1)]\ dx\ dy$$

S = bottom + top + sides

$$\unicode{x222F}_S \langle 0,0,R \rangle \cdot \hat n \ dS = \iint_{top} + \iint_{bottom} + \iint_{sides}$$


### Top: graph $z = z_2(x,y)$

Because

$\hat n \ dS = \langle -\frac{\partial z_2}{\partial x}, -\frac{\partial z_2}{\partial y}, 1 \rangle\ dx\ dy$ 

$\langle 0,0,R \rangle \cdot \hat n\ dS = R\ dx\ dy$ and $U$ is the area projected on xy plane

$$\iint_{top} \langle 0,0,R \rangle \cdot \hat n\ dS = \iint_{top} R\ dx\ dy = \iint_U R(x,y,z_2)\ dx\ dy$$


### Bottom: graph $z = z_1(x,y)$

$\hat n \ dS = \langle \frac{\partial z_1}{\partial x}, \frac{\partial z_1}{\partial y}, -1 \rangle\ dx\ dy$ 

$\langle 0,0,R \rangle \cdot \hat n\ dS = -R\ dx\ dy$

$$\iint_{bottom} \langle 0,0,R \rangle \cdot \hat n\ dS = \iint_{bottom} -R dx\ dy = \iint_U -R(x,y,z_1) dx\ dy$$

### Sides are vertical

$\langle 0,0,R \rangle$ is tangent to the sides, flux through sides = 0


Recap:

$$\iiint_D R_z\ dV = \unicode{x222F}_{top+bot+sides} \langle 0,0,R \rangle \cdot \hat n\ dS$$

If D is not vertically simple, decompose it into simple pieces

## Diffusion Equation (Heat equation)

governs motion of smoke in immobile air, dye in immobile solution (no stiring)

$u =$ concentration (or temperature) at a given point, depends on space and time = $u(x,y,z,t)$ 

$$\frac{\partial u}{\partial t} = k \nabla^2 u = k\ \nabla \cdot \nabla u = k(\frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} + \frac{\partial^2 u}{\partial z^2})$$ 

$\nabla^2 u$ is Laplacian

$\nabla \cdot \nabla u$ is $div(\nabla u)$

$\vec F = \text{flow of smoke}$

- Physics (+ common sense): smoke flows from high concentration towards low concentration region. So $\vec F$ is directed along $-\nabla u$. In fact, $\vec F = -k \nabla u$


- Relate $\vec F$ and $\frac{\partial u}{\partial t}$? Math - divergence


Flux out of D through S:

$$\unicode{x222F}_S \vec F \cdot \hat n\ dS = \text{amount of smoke leaving S per unit time} = \frac{-d}{dt} \iiint_D u\ dV $$

According to Divergence theorem

$$\unicode{x222F}_S \vec F \cdot \hat n\ dS = \iiint_D div(\vec F)\ dV = \frac{-d}{dt} \iiint_D u\ dV = -\iiint_D \frac{\partial u}{\partial t}\ dV$$

The last step works because we can think of $\frac{d}{dt} \iiint_D u\ dV$ as differentiating sum $\frac{d}{dt} (\sum_i u(x_i, y_i, z_i, t) \Delta v_i)$, which is equal to differentiate each terms. $\sum_i \frac{\partial }{\partial t} u(x_i,y_i,z_i,t) \Delta v_i$

The derivative of the sum is equal to the sum of derivative. Also the variable of derivative is not the same as variable of integration

For any region D, 

$$\frac{1}{Vol(D)} \iiint_D div \vec F\ dV = \frac{1}{Vol(D)} \iiint_D -\frac{\partial u}{\partial t}\ dV$$

Average of $div \vec F$ in D = average of $-\frac{\partial u}{\partial t}$ in D

$$div \vec F = -\frac{\partial u}{\partial t}$$

how much we are expanding equal to how much we are losing. This relates $\vec F$ and $\frac{\partial u}{\partial t}$
