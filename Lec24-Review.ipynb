# Review

More about validity of Green's theorem

We've seen Green's theorem

$$\oint_c \vec F \cdot \hat T\ ds = \iint_R curl(\vec F)\ dA$$

$$\oint_c \vec F \cdot \hat n\ ds = \iint_R div(\vec F)\ dA$$

Only works if $\vec F$ (and derivative) defined everywhere in R

Example: $$\vec F = \frac{-y\hat i + x\hat j}{x^2 + y^2}$$

$\vec F$ is not defined at origin but defined everywhere else where $curl(\vec F) = 0$ 

If closed curve doesn't include origin, then 

$$\oint_c \vec F \cdot \hat T\ ds = \iint_R curl(\vec F)\ dA = 0$$

If enclosed curve includes origin, then we can't use Green directly. Let say we have enclose curve C' anticlockwise includes origin, and C'' is a circle anticlockwise include origin

$$\oint_{C'} \vec F \cdot d\vec r - \oint_{C''} \vec F \cdot d\vec r = \iint_R curl(\vec F)\ dA = 0$$

Why?

Because we can see this as outer C' connects to inner C'' and then connect back to C'. Then it becomes a curve excludes origin.

Total line integral = $\oint_{C'} - \oint_{C''}$. For the connected line, one goes in, one goes out, so they cancel out

We can apply this trick to explain what is happening


# Simply Connected Region

A connected region in the plane is simply connected if the interior of any closed curve in R is also contained in R

That means the region R does not have any holes inside

If domain where $\vec F$ is defined and differentiable is simply connected. then can always apply Green's theorem

So the example above $\vec F = \frac{-y\hat i + x\hat j}{x^2 + y^2}$ whose domain is plane minus origin is NOT simply connected

If $curl(\vec F) = 0$ and domain where $\vec F$ defined is simply connected, then $\vec F$ is conservative and is a gradient field

$$\oint \vec F \cdot d\vec r = \int_c M\ dx + N\ dy$$

$$\int_c \vec F \cdot \hat n\ ds = \int_c -N\ dx + M\ dy$$
