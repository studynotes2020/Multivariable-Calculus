# Double Integrals

Remember: function of 1 variable $\int_a^b f(x)\ dx $ is the area below graph of f over [a, b]

Double integral is the volume below graph $z = f(x,y)$ over a region R in xy plane

$$\iint_R f(x,y)\ dA$$

Cut into many small pieces with small area $\Delta A$, multiplies with the height which is $f(x, y)$, sum all those small volumes up. Integrals will be defined as a limit where the pieces become smaller and smaller

Definition: Cut R into small pieces of area $\Delta A_i$

$$\sum_i f(x_i, y_i) \cdot \Delta A_i$$

Take the limit as $\Delta A_i \to 0$, get $\iint$

To compute $\iint_R f(x,y) \ dA$, let $S(x)$ is the area of slice by the plane parallel to yz plane. Take the slice, move it from min x to max x. So volume is 

$$\int_{x_{min}}^{x_{max}} S(x)\ dx$$


For given x, $$S(x) = \int_{y_{min}(x)}^{y_{max}(x)} f(x,y)\ dy$$

Note that the value of y depend on x

$$\iint_R f(x, y)\ dA = \int_{x_{min}}^{x_{max}} \int_{y_{min}(x)}^{y_{max}(x)} f(x,y)\ dy\ dx$$

This is called an ITERATED INTEGRAL because iterated twice the process of taking an integral

The bounds $x_{min}$ and $x_{max}$ are just numbers. They are the first value and last value of x to slice. Any region between them, we would like to integrate

However, the inner integral depends on value of x. Because region between $x_{min}$ and $x_{max}$, y depends on x

### Example 1

Integrate $z = 1 - x^2 - y^2$

region $0 \leq x \leq 1$, $0 \leq y \leq 1$

$$\int_{0}^{1} \int_{0}^{1} 1 - x^2 - y^2\ dy\ dx$$

Inner integral: $$ \int_{0}^{1} 1 - x^2 - y^2\ dy = [y - x^2y - \frac{y^3}{3}]_0^1 = 1 - x^2 - \frac13 = \frac23 - x^2$$

Outer integral: $$ \int_{0}^{1} (\frac23 - x^2)\ dx = [\frac23 x - \frac{1}{3} x^3]_0^1 = \frac13$$


$dA = dy\ dx$ because $\Delta A_i = \Delta x \Delta y$ for small triangle 


### Example 2

Same function, R is quarter disk, $x^2 + y^2 \leq 1, x \geq 0, y \geq 0$

For given x, the range of y is from 0 to $y = \sqrt{1 - x^2}$

$$\iint_R 1 - x^2 - y^2\ dA = \int_0^1 \int_0^{\sqrt{1 - x^2}} (1 - x^2 - y^2)\ dy\ dx$$


```julia
using Plots
plot(x->sqrt(1-x^2),0, 1)
```

Inner: 

$$\begin{align}
\int_0^{\sqrt{1 - x^2}} (1 - x^2 - y^2)\ dy & = [y - x^2y - \frac{y^3}{3}]_0^{\sqrt{1 - x^2}} \\
& = (1 - x^2)\sqrt{1 - x^2} - \frac{(1-x^2)^{3/2}}{3} \\
& = \frac23 (1 - x^2)^{3/2}
\end{align}$$

Outer

$$\int_0^1 \frac23 (1 - x^2)^{3/2}\ dx = \int_0^{\pi/2} \frac23 \cos^3 \theta \cdot \cos\theta\ d\theta$$

$x = \sin\theta, (1-x^2)^{1/2} = \cos\theta, dx = \cos\theta \ d\theta$

$$= \frac23 \int_0^{\pi/2} \cos^4 \theta\ d\theta = \frac23 \int_0^{\pi/2} (\frac{1 + \cos (2\theta)}{2})^2\ d\theta$$

$$= \frac23 \int_0^{\pi/2} (\frac14 + \frac12 \cos(2\theta) + \frac14 \cos^2(2\theta)\ d\theta)$$

Use the relation between cos square and cos 2 theta, in the end the result is $\frac{\pi}{8}$

There is a easier way to compute this, which is using polar coordinates


## Exchanging order of integration

We can exchange the order but be careful on the bound

Example 1: $$\int_0^1 \int_0^2 dx\ dy = \int_0^2 \int_0^1 dy\ dx$$

This is a rectangle. Here is work, but not for others

Example 2: $$\int_0^1 \int_x^{\sqrt{x}} \frac{e^y}{y} dy\ dx$$

This is an interesting example, because we cannot integrate $\int \frac{e^y}{y} dy$. So, we need to switch the order


```julia
plot(x->x, 0, 2)
plot!(x->sqrt(x), 0, 2)

# Want to integrate between x = 0 to 1
```

Take a horizontal slice where y enters with $x = y^2$ and leaves at $x = y$

Start and end points of y are from 0 to 1

$$\int_0^1 \int_{y^2}^{y} \frac{e^y}{y} dx\ dy$$

After exchanging the order, it can be integrated

Inner: $$\int_{y^2}^{y} \frac{e^y}{y} dx = [\frac{e^y}{y} x]_{y^2}^y = e^y - e^y y$$

Outer: $$\int_0^1 e^y - e^y y \ dy = [-y e^y + 2e^y]_0^1 = e -2$$

$(-y e^y)' = -e^y - y e^y$
