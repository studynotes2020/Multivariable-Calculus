# Unit 1

### Vector

- Dot product

$$\vec A \cdot \vec B = \sum a_i b_i = |\vec A| |\vec B| \cos\theta$$

- Cross product

$$\vec A \times \vec B$$

Magnitude is the area formed by vector A and B. The direction is perpendicular to vector A and B

- Equation of plane

$ax + by + cz = d$, $\langle a,b,c \rangle = $ normal vector

- Equation of lines

parametric $\begin{cases}
x = x_0 + at \\
y = y_0 + bt \\
z = z_0 + ct
\end{cases}$

$x_0, y_0, z_0$ are points on the line. $a, b, c$ is vector parallel to the line

- Parametric equation of curves

Decompose position vector into sum. For example give $x^2 + y^2 = a^2$ and we have expression of z in x and y, $z = c_1x + c_2y + d$. We can express x and y in a single variable t, then z

$\begin{cases}
x = a\cos t \\
y = a\sin t \\
z = ac_1\cos t + ac_2\sin t + d
\end{cases}$

- velocity

$\vec v = \frac{d\vec r}{dt},\quad speed = |\vec v|$

$\vec v = \hat T \frac{ds}{dt}$

where $\hat T$ is unit vector, $\frac{ds}{dt} = |\vec v|$ is speed

Acceleration $\vec a = \frac{d\vec v}{dt}$


- matrices, determinants, linear systems

3x3 linear system can be writeen in matrix form AX = B

A is 3x3 matrix. X is column vector, B is column vector

Inverting 2x2 and 3x3 matrices. Must be a square matrix

Matrix -> minor, entries are 2x2 determinants formed by deleting 1 row and 1 column -> signs of cofactors -> transpose -> divide by det(A)

A invertible <-> det(A) not zero. Then $Ax = B \implies x = A^{-1}B$

Otherwise $Ax = B$ has either no solution or infinitely many solutions

Geometrically, determinant is zero means all 3 planes are parallel to a same vector

If Ax = 0 then x = 0 is always a solution (trivial solution)

# Unit 2

Viewing f(x,y) as a graph, contour plot

When we have two variables, then the graph is a surface given by the equation z = f(x,y)

Contour plot tells us what are each levels (heights) of the graph. The number label on the curve is the value of the function of that height


- Partial derivative

$$f_x = \frac{\partial f}{\partial x}, \quad f_y = \frac{\partial f}{\partial y}$$

linear approximation: $\Delta f \approx f_x \Delta x + f_y\Delta y$. This also tells us about tangent plane to the graph

- Differentials, chain rules

$$df = f_x\ dx + f_y\ dy$$

If x = x(t), y = y(t), then 

$$\frac{df}{dt} = f_x \frac{dx}{dt} + f_y \frac{dy}{dt}$$


- gradient vector

$$\nabla f = \langle f_x, f_y \rangle$$

The gradient vector is always perpendicular to the contour plot and points towards higher ground

Directional derivative for unit $\hat u$

$$\frac{df}{ds|_{\hat u}} = \nabla f \cdot \hat u$$

To calculate the flux through a curve, we need to know its normal vector

Let say, surface/curve f = c, then its normal vector is $\nabla f = \vec N$

Convert to unit normal vector $\frac{\nabla f}{|\nabla f|} = \hat n$

- max/min problems

critical points: $\nabla f = 0$

Second derivative to determine it is local min or local max or saddle

Also need to check the value of f at the boundary or at infinity


- max/min f with non-independent variables (g = c)

then we use Lagrange multipliers

$\begin{cases}
\nabla f = \lambda \nabla g \\
g = c
\end{cases}$

Second derivative test does not apply here!


- Constrained partial derivatives

g(x,y,z) = c

Formal partial: $\frac{\partial f}{\partial x}$ x varies, y and z held constant

$$(\frac{\partial f}{\partial x})_y$$

y held constant, x varies and z as a depends on x and y

$$(\frac{\partial f}{\partial x})_z$$

means y depends on x, z
