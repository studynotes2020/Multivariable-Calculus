# Equations of Lines

We have seen a line as an intersection of 2 planes

Another way,a line is a trajectory of a moving point. This is called parametric equation

Example: Line through $Q_0 = (-1,2,2), Q_1=(1,3,-1)$

$Q(t)$ is a moving point. At t = 0, the point at Q0, at time = 1, the point at Q1. It moves at constant speed on the line

What is the position at time t, $Q(t)$?

$\overrightarrow {Q_0 Q(t)} = t \cdot \overrightarrow {Q_0Q_1} = t \cdot \langle 2,1,-3 \rangle$

$Q(t) = (x(t), y(t), z(t))$

$Q(t) - Q_0 = t \cdot \overrightarrow {Q_0Q_1}$ 

$\begin{cases} x(t) + 1 = 2t \\
y(t) - 2 = t \\
z(t) - 2 = -3t
\end{cases}$

$\begin{cases} x(t) = -1 + 2t \\
y(t) = 2 + t \\
z(t) = 2 -3t
\end{cases}$

$Q(t) = Q_0 + t \cdot \overrightarrow {Q_0Q_1} $



## Application

Intersection with a plane?

Consider the plane $x + 2y + 4z = 7$. Q0 = (-1,2,2), Q1 = (1,3,-1) 

Where does the line through $Q_0, Q_1$ intersect the plane?

Substitute $Q_0 = (-1,2,2) \implies x + 2y + 4z = (-1) + 2.2 + 4.2 = 11 > 7$. $Q_0$ is not in the plane

Substitute $Q_1 = (1,3,-1) \implies x + 2y + 4z = 1 + 2.3 + 4.(-1) = 3 < 7$. $Q_1$ is not in the plane

The equation when changing 11 to 3, will need to cross 7

$Q_0, Q_1$ are on the opposite sides of the plane

What about $Q(t)$?

$x(t) + 2y(t) + 4z(t) = -1 + 2t + 2(2 + t) + 4(2 -3t) = -8t + 11$

$Q(t)$ is in the plane when $-8t + 11 = 7 \implies t = \frac12$

Then $Q(\frac12) = (0,\frac52, \frac12)$. The line intersects the plane at $(0,\frac52, \frac12)$

Note: if line is in the plane then plugging $(x(t), y(t), z(t))$ into equations always get 7

If line is parallel to the plane, we get another constant

More generally, parametric equations for arbitrary motion in the plane or in space!

## Cycloid

Let say there is a wheel of radius a rolling on a horizontal ground (x-axis) . P is a point on the rim, starts at origin. The cycloid is a trajectory of the moving point. Can we find the position of P as a function of time? What is $(x(t), y(t))$?

The trajectory is the same for whatever speed of the wheel. What matter is how far the wheel has moved. We can choose different variables to parameterize. Here, the simplest is to parameterize with angle.

Motion is a function of the angle theta by which the wheel has rotated.

Let A be the point of wheel touching the ground, and B is the center of the wheel

$\vec{OP} = \vec{OA} + \vec{AB} + \vec{BP}$

$\vec{OA} = \langle a\theta, 0 \rangle$, theta in radians

$\vec{AB} = \langle 0, a \rangle$

$\vec{BP}$ has length a ($|\vec{BP}| = a$) and angle $\theta$ with vertical

$\vec{BP} = \langle -a\sin\theta, -a\cos\theta \rangle$

$\vec{OP} = \langle a\theta - a\sin\theta, a - a\cos\theta \rangle = \langle x(\theta), y(\theta) \rangle$

Question: What happens near the bottom?

Answer: Take length unit as radius: a = 1

$\begin{cases} x(\theta) = \theta - \sin\theta \\
y(\theta) = 1 - \cos\theta
\end{cases}$

When $\theta$ is very small $\sin\theta \approx \theta , \cos\theta \approx 1$

Then we get $x(\theta) = 0, y(\theta) = 0$, no going anywhere. We need better approximation -> Taylor expansion


### Taylor Appoximation

For t is small, $f(t) \approx f(0) + t f'(0) + \frac{t^2}{2}f''(0) + \frac{t^3}{6}f'''(0) + ...$

$\sin\theta \approx \theta - \frac{\theta^3}{6}$

$\cos\theta \approx 1 - \frac{\theta^2}{2}$

$x(\theta) \approx \theta - (\theta - \frac{\theta^3}{6}) \approx \frac{\theta^3}{6}$

$y(\theta) \approx 1 - (1 - \frac{\theta^2}{2}) \approx \frac{\theta^2}{2}$

Hence $|x(\theta)| << |y(\theta)|$

$$Slope = \frac{y}{x} \approx \frac{\frac{\theta^2}{2}}{\frac{\theta^3}{6}} = \frac{3}{\theta} \to \infty$$

when $\theta \to 0$ which means the slope is infinity when $\theta$ is small.

So the instantaneous motion at 0 angle is moving down and up
