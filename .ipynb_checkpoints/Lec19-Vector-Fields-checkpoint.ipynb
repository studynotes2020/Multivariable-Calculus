# Vector Fields

$$\vec F = M \hat i + N \hat j$$

$M, N$ are functions of x, y

At each point (x, y), $\vec F$ is a vector that depends on (x, y)

Examples: velocity in a fluid $\vec v$ or force field $\vec F$

Examples: $\vec F = 2 \hat i + \hat j$. This means independent of x and y. Every points in xy plane have the same vector. 

Examples: $\vec F = x \hat i$. Every vector points to the right, its magnitude depends on x. For -x, vectors point to the left

Examples: $\vec F = x\hat i + y\hat j$. Vectors point radially away from the origin and its magnitude increases with the distance from the origin

Examples: $\vec F = -y\hat i + x\hat j$. $\langle x, y \rangle$ rotate 90 anticlockwise is $\langle -y, x \rangle$. So vectors in xy plane are pointing to 90 anticlockwise direction. Moving away from the origin, the vector magnitudes become bigger. Velocity field for uniform rotation at unit angular velocity


## Work and Line Integrals

Work = force * distance = $\vec F \cdot \Delta \vec r$

How about a trajectory where force are acting differently along the trajectory?

In that case, we divide the trajectory into many $\Delta r$ then take its dot product with force, sum all the work done. So it will be an integral

Along a trajectory C, work adds up to $$W = \int_c \vec F \cdot d\vec r$$

This is the limit, $$\lim\limits_{\Delta r_i \to 0} \sum_i \vec F \cdot \Delta \vec r_i = \lim\limits_{\Delta t \to 0} \sum_i \vec F \cdot \frac{\Delta \vec r}{\Delta t} \Delta t$$

$\frac{\Delta \vec r}{\Delta t}$ is the velocity vector $\frac{d\vec r}{dt}$. Then we can compute W

$$\int_c \vec F \cdot d\vec r = \int_{t_1}^{t_2} \vec F \cdot \frac{d\vec r}{dt}\ dt$$

F at the point on the trajectory at time t. It depends on x, y before it depends on t

Example: $\vec F = -y\hat i + x\hat j$ The trajectory is along a parametric curve $x = t, y = t^2, 0 \leq t \leq 1$

We have the force and trajectory, we want to find the work done

$$\int_c \vec F \cdot d\vec r = \int_0^1 \vec F \cdot \frac{d\vec r}{dt}\ dt = \int_0^1 \langle -t^2, t \rangle \cdot \langle 1, 2t \rangle \ dt$$

because $\vec F = \langle -y, x \rangle = \langle t^2, t \rangle$ and $x' = 1, y' = 2t$

$$\int_0^1 (-t^2 + 2t^2)\ dt = \int_0^1 t^2 \ dt = \frac13$$

Currently, we don't know yet how to integrate F with vector r. So we do this method

### Another way

$\vec F = \langle M, N \rangle$, $d\vec r = \langle dx, dy \rangle$, then $\vec F \cdot d\vec r = M\ dx + N\ dy$

$$\int_c \vec F \cdot d\vec r = \int_c M\ dx + N\ dy$$

Method to evaluate -> Express x, y in terms of a single variable and substitute

$$\int_c \vec F \cdot d\vec r = \int_c -y\ dx + x\ dy$$

Because $x = t, y = t^2 \implies dx = dt, dy = 2t\ dt$

$$= \int_c -t^2\ dt + t.2t\ dt$$

$$= \int_0^1 t^2\ dt = \frac13$$

Note: $\int_c \vec F \cdot d\vec r$ depends on the trajectory $C$ but not on parameterization

e.g. could do $x = \sin\theta, y = \sin^2\theta, 0 \leq \theta \leq \frac{\pi}{2}$ (Not practical)

If we can express y in term of x, then we express all in term of x and dx



## Geometric approach

What is $\Delta \vec r$?

It is going in the same direction as unit vector T. Its length is the arc length of the trajectory $\Delta s$

$$d\vec r = \langle dx, dy \rangle = \hat T\ ds$$

Similarly: $$\frac{d\vec r}{dt} = \langle \frac{dx}{dt}, \frac{dy}{dt} \rangle = \hat T \frac{ds}{dt}$$

$\frac{ds}{dt}$ is the velocity

$$\int_c \vec F \cdot d\vec r = \int_c M\ dx + N\ dy = \int_c \vec F \cdot \hat T\ ds$$

Force projected on the tangent of the curve then integrate

Example: $C$ is circle of radius $a$ at origin, counterclockwise, $\vec F = x\hat i + y\hat j$. So the trajectory is a circle and forces pointing radially away from the origin. Because the force is perpendicular to the motion, so the work will be zero

$\vec F \bot \hat T$, so $\vec F \cdot \hat T = 0$, $\int_c \vec F \cdot \hat T \ ds = 0$

Example: Same $C$, but $\vec F = -y\hat i + x\hat j$. 

Then $\vec F$ parallel $\hat T$, $\vec F \cdot \hat T = |\vec F| = a$

$$\int_c \vec F \cdot d\vec r = \int_c a\ ds = a \int_c ds = a.2\pi a = 2\pi a^2$$

Because trajectory is a circle, hence $2\pi a$

Or computing $x = a\cos\theta, y = a\sin\theta, 0 \leq \theta \leq 2\pi$

$$\int_c -y\ dx + x\ dy = \int_c -(a\sin\theta)(-a\sin\theta \ d\theta) + (a\cos\theta)(a\cos\theta \ d\theta)$$

$$ = \int_0^{2\pi} a^2(\sin^2\theta + \cos^2\theta) \ d\theta = 2\pi a^2$$
