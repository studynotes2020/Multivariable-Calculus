# Flux

Flux is another kind of line integral. Given C plane curve, $\vec F$ vector field, Flux of $\vec F$ across C is 

$$\int_c \vec F \cdot \hat n\ ds$$

$\hat n$ is unit normal vector to C, 90 clockwise from $\hat T$

There are two normal vectors pointing different directions. We have to choose a convention. The convention is when we walk on the curve, the normal vector $\hat n$ points towards right hand side

If break C into small pieces, $\Delta s$, then the flux will be the limit of s going zero, of the sum of vector field dot normal vector and multiply $\Delta s$

$$Flux = \lim\limits_{\Delta s \to 0} \sum \vec F \cdot \hat n (\Delta s)$$ 

$$Work = \int_c \vec F \cdot d\vec r = \int_c \vec F \cdot \hat T\ ds$$

Work is about summing tangential component of $\vec F$

Rougly speaking, work measure how much vector field goes along the curve. Flux measure how much vector field goes perpendicular to the curve. Counting positively what goes to right and negative for those go the left

Flux is about summing normal component of $\vec F$


## Interpretation

for $\vec F$ a velocity field, flux measures how much fluid passes through C per unit time. Imagine fluids flow across curve C in unit time and C is chopped in small piece s

What passes through a portion of C in unit time?

Area = base x height = $\Delta s \times \vec F \cdot \hat n$

where $\hat n$ in normal to $\Delta s$. time and $\Delta s$ has to be sufficienly small so that it is a straight line for the equation to be valid

What flows across C from left to right is counted positively while right to left is counted negatively

### Example:

C is a circle of radius a at origin, counterclockwise

$\vec F = x\hat i + y\hat j$

Along C, $\vec F$ is parallel to $\hat n$, $\vec F \cdot \hat n = |\vec F| = a$, if the radius of the circle is a

$$\int_c \vec F \cdot \hat n\ ds = \int_c a\ ds = a \cdot length(C) = 2\pi a^2$$

Same C, but different $\vec F = \langle -y, x \rangle$. Now the vector field is just tangent to the circle C, then $\vec F \cdot \hat n = 0$, Flux = 0




## Calculation using components

Remember: $d\vec r = \hat T\ ds = \langle dx, dy \rangle$

$\hat n$ is $\hat T$ rotated 90 clockwise $\hat n\ ds = \langle dy, -dx \rangle$

Let say a small piece of C

$$\hat T \Delta s = \Delta \vec r = \langle \Delta x, \Delta y \rangle$$

$$\hat n \Delta s = \langle \Delta y, -\Delta x \rangle$$

If $\vec F = \langle M, N \rangle$, then $\int_c \vec F \cdot \hat n\ ds = \int_c \langle M, N \rangle \cdot \langle dy, -dx \rangle = \int_c -N\ dx + M\ dy$



## Green's Theorem for flux

If C encloses a region R counterclockwise, and $\vec F = \langle P, Q \rangle$ defined in R, then $\oint \vec F \cdot \hat n\ ds = \iint_R div(\vec F)\ dA$. Divergence of $\vec F$

$$div \langle P, Q \rangle = P_x + Q_y$$

For counterclockwise, the flux flows out of R through C

This is called Green's theorem in normal form (versus Green's in tangential form)

Proof: $$\oint_c -Q\ dx + P\ dy = \iint _R (P_x + Q_y)\ dA$$

Let $M = -Q, N = P$

For LHS, substitute M, N and using Green's for tangential form from last lecture 

$$\oint_c -Q\ dx + P\ dy = \oint_c M\ dx + N\ dy = \iint _R (N_x - M_y)\ dA$$

For RHS

$$\iint _R (P_x + Q_y)\ dA = \iint _R (N_x - M_y)\ dA$$

This proves two quantities are equal by just renaming the components. So actually both are the same theorem

But the way we think about it is different. One computes work done by a force along a closed curve. The other one computes the flux of a velocity field flows out of the region

If $\vec F = \langle M, N \rangle$, then $\oint_c -N\ dx + M\ dy = \iint _R (M_x + N_y)\ dA$



### Example

$\vec F = x\hat i + y\hat j$, C is a circle of radius a

$$div \vec F = \frac{\partial x}{\partial x} + \frac{\partial y}{\partial y} = 1 + 1 = 2$$

Green theorem tells us 

$$\oint \vec F \cdot \hat n\ ds = \iint_R 2\ dA = 2\pi a^2$$

because we know area of a circle is $\pi a^2$. From the calculation, it doesn't matter where is the origin


## Interpretation of divergence $\vec F$

- It measures how much the flow is expanding. If the thing is contracting, then it has negative divergence

- Source rate -> amount of fluid added to the system per unit time, per unit area
