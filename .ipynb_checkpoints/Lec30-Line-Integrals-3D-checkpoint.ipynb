# Diffusion Equation

$u$ is concentration of substance. $\vec F$ is flow of substance

1. Physics: The concentration goes from high to low concentration. $\vec F = -k \nabla u$


2. Derive from Divergence theorem: $$div \vec F = -\frac{\partial u}{\partial t}$$

Hence $$\frac{\partial u}{\partial t} = -div \vec F = k \cdot div(\nabla u) = k \nabla^2 u$$

We got diffusion equation $$\frac{\partial u}{\partial t} = k \nabla^2 u$$

# Line Integrals in 3D

Vector field (force) $\vec F = P\hat i + Q\hat j + R\hat k$

Curve C in space: 

$$Work = \int_c \vec F \cdot d\vec r = \int_c P\ dx + Q\ dy + R\ dz$$

where $d\vec r = \langle dx, dy, dz \rangle$

Evaluate: parameterize C, express x, y, z and dx, dy, dz in terms of the parameter

### Example 1   
$\vec F = \langle yz, xz, xy \rangle$, where $x = t^3, y = t^2, z = t, 0 \leq t \leq 1$

$dx = 3t^2\ dt, dy = 2t\ dt, dz = dt$

$$\int_c \vec F \cdot d\vec r = \int_c yz\ dx + xz\ dy + xy\ dz = \int_c t^3 . 3t^2\ dt + t^4.2t\ dt + t^5\ dt$$

$$= \int_0^1 6t^5\ dt = [t^6]_0^1 = 1$$

Instead of using t, here we could use z as the parameter. In other question, we could have used angle

### Example 2

C' = C1 + C2 + C3 where C1 moves from (0,0,0) to (1,0,0), C2 moves from (1,0,0) to (1,1,), C3 moves from (1,1,0) to (1,1,1)

$$\int_{C'} = \int_{C1} + \int_{C2} + \int_{C3}$$

For C1, C2, they are in the xy plane. $z = 0, dz = 0$

$$\int_{C1} yz\ dx + xz\ dy + xy\ dz = 0$$

For C3, $x = 1, y = 1, dx = 0, dy = 0$ and z varies from 0 to 1

$$\int_{C3} \vec F \cdot d\vec r = \int_{C3} xy\ dz = \int_0^1 dz = 1$$

$$\int_{C'} = \int_{C1} + \int_{C2} + \int_{C3} = 0 + 0 + 1 = 1$$

In fact, $\vec F$ is conservative. C in example 1 and C' in example 2 both go from (0,0,0) to (1,1,1). Therefore

$$\int_C \vec F \cdot d\vec r = \int_{C'} \vec F \cdot d\vec r$$

Given $f(x,y,z) = xyz$

$$\vec F = \nabla f(x,y,z) = \langle yz, xz, xy \rangle$$

Knowing this: Fundamental Theorem of Calculus for line integrals

$$\int_C \nabla f \cdot d\vec r = f(p_1) - f(p_0)$$

here $f(1,1,1) - f(0,0,0) = 1 - 0 = 1$

## Test for gradient fields

To know if this is gradient field, we want to check if this is true

$\vec F = \langle P,Q,R \rangle = \langle f_x, f_y, f_z \rangle?$

If so then 

$P_y = f_{xy} = f_{yx} = Q_x$ and 

$P_z = f_{xz} = f_{zx} = R_x$ and 

$Q_z = f_{yz} = f_{zy} = R_y$

critetion: $\vec F = \langle P, Q, R \rangle$ (defined in a simply connected region ) is a gradient field $\iff$

$\begin{cases}
P_y = Q_x \\
P_z = R_x \\
Q_z = R_y
\end{cases}$

$P\ dx + Q\ dy + R\ dz$ is exact differential, $df$ with the same condition

Example: for which a and b is $axy\ dx + (x^2 + z^3)\ dy + (byz^2 - 4z^3)\ dz$ exact differential?

$P_y = ax = 2x = Q_x \implies a = 2$

$P_z = 0 = 0 = R_x$

$Q_z = 3z^2 = bz^2 = R_y \implies b = 3$

Any value other than a=2 and b=3, we need to use parameterize method. For a=2, b=3, we show the method of computing below



### Find potential

1. $f(x_1, y_1, z_1) = \int_c \vec F \cdot d\vec r$ (+ constant) where c from (0,0,0) to $(x_1, y_1, z_1)$. First goes along x-axis, then goes parallel to y-axis, finally goes along z-axis to $(x_1, y_1, z_1)$. Then sum the results of these three line integrals. This works the same as in 2D


2. Antiderivatives. We want to solve 

$P = f_x = 2xy$ 

$Q = f_y = x^2 + z^3$

$R = f_z = 3yz^2 - 4z^3$

Integrate the first equation

$\int f_x\ dx \implies f = x^2 y + g(y,z)$

Partial derivative wrt y, $f_y = x^2 + g_y$, Compare with $f_y$, we get $g_y = z^3$

Integrate g wrt y

$\int g_y\ dy \implies g = yz^3 + h(z)$. Plug in back to f

$f = x^2 y + yz^3 + h(z)$. Partial derivative wrt z, $f_z = 3yz^2 + h'$

Compare with third equation, $h'(z) = -4z^3 \implies h = -z^4 + C$

This C is a true number that doesn't depend on x,y,z

$f = x^2 y + yz^3 - z^4 + C$



## Curl in 3D

If $\vec F = P\hat i + Q\hat j$, $curl \vec F = (R_y - Q_z)\hat i + (P_z - R_x)\hat j + (Q_x - P_y)\hat k$

If $\vec F$ is defined in a simply connected region, $\vec F$ is conservative $\iff$ $curl \vec F = 0$

Use $\nabla$ and $\times$ to remember the theorem

$$\nabla = \langle \frac{\partial }{\partial x}, \frac{\partial }{\partial y}, \frac{\partial }{\partial z} \rangle$$

We have seen: 

$$\nabla f = \langle \frac{\partial }{\partial x}, \frac{\partial }{\partial y}, \frac{\partial }{\partial z} \rangle$$

$$\nabla \cdot \langle P,Q,R \rangle = \frac{\partial P}{\partial x} +  \frac{\partial Q}{\partial y} + \frac{\partial R}{\partial z} = div\vec F$$

The new one

$$\nabla \times \vec F = \begin{vmatrix}\hat i & \hat j & \hat k \\
\frac{\partial }{\partial x} & \frac{\partial }{\partial y} & \frac{\partial }{\partial z} \\
P & Q & R
\end{vmatrix} = (\frac{\partial R}{\partial y} - \frac{\partial Q}{\partial z})\hat i - (\frac{\partial R}{\partial x} - \frac{\partial P}{\partial z})\hat j + (\frac{\partial Q}{\partial x} - \frac{\partial P}{\partial y})\hat k = curl\vec F$$


## Geometrically

Curl measures rotation component in a velocity field

Given a rotation about z-axis at angular velocity $w$

$$\vec v = \langle -wy, wx, 0 \rangle$$

After calculation, we will get curl is 2w

$$curl \vec v = \nabla \times \vec v = 2w\hat k$$

Curl gives angular velocity of rotation (with a factor of 2), and the direction of the axis of rotation. Here, it is rotating about the vertical axis
