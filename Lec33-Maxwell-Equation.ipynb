# Curl relation with Physics

What curl says about force field?

Recall: $\vec v$ velocity field -> curl $\vec v$ measure 2 * angular velocity vector (rotation path of the motion)


### Uniform rotation about z-axis

$\vec v = w(-y\hat i + x\hat j)$

$\nabla \times \vec v = 2w \hat k$

What does the curl for a force field means?

A small piece of mass $\Delta m$ with force applied to it, $\vec F \Delta m$. The mass is sitting at position vector $\vec r$ from a reference frame.

Torque is a vector $\vec \tau = \vec r \times \vec F \Delta m$

For translation, force/mass = acceleration = derivative of velocity. 

For rotation: torque/moment of inertia = angular acceleration = derivative of angular velocity

So, curl is an operation on velocity that gives angular velocity for its rotation effects

Curl of an acceleration field give the angular acceleration in the rotation part of the acceleration effects

Curl for a force field measures the torque per unit of moment of inertia. If you leave a small mass in a force field, curl measures how much the mass will spin. Conversely, a force field with no curl, does not generate any rotation motion.

Consequence: if force $\vec F$ derives from a potential, then $curl \vec F = 0$. $\vec F$ does not induced any rotation motion. For example, gravitational attraction cannot causes the earth to spin faster or slower. The reason the earth is spinning is because it was formed spinning. It did not start spinning because the gravitational effect


### Electric and magnetic fields

Maxwell equation governs how electric and magnetic fields behave, how they are caused by electric charges and their motions

Electric field $\vec E$ is a vector field. A particle with an electric charge, there will be an electric force which is qE exerted on that particle sitting inside the electric field

$$\vec F = q\vec E$$

$$\vec F = q\vec v \times \vec B$$

Maxwell equation tells us div and curl of $\vec E$

Gauss-Coulomb Law: $\vec \nabla \cdot \vec E = \frac{\rho}{\epsilon_0}$. $\rho$ is electric charge density, $\epsilon_0$ is a constant. This tells us divergence of E is caused by the presence of electric charge

Any closed surface, electric field out of this surface, we want to find the electric field out of this surface

$$\iint_S \vec E \cdot d\vec S = \iiint_D div \vec E\ dV = \iiint_D \frac{1}{\epsilon_0} \rho\ dV = \frac{Q}{\epsilon_0}$$

Integrate electric charge density over entire volume, get total amount of total charge inside D. This tells us how the electric charge inside S influence the electric field around them


### Faraday's law

$$\nabla \times \vec E = \frac{\partial \vec B}{\partial t}$$

Electric field deriving from electric potential becomes false when you have variable magnetic field. Or a varying magnetic field can generate electric field. For example transformer generate varying magnetic field thus generate electric field

From Stokes theorem, we want to calculate the electric generated from magnetic field

$$\oint_c \vec E \cdot d\vec r = \iint_S (\nabla \times \vec E)\cdot d\vec S = \iint_S \frac{\partial \vec B}{\partial t} \cdot d\vec S$$

Remark: Divergence of magnetic field is zero. 

$$\nabla \cdot \vec B = 0$$. 

$$\nabla \times \vec B = \mu_0 \vec J + \epsilon_0 \mu_0 \frac{\partial \vec E}{\partial t}$$

J is current density, measures the flow of electrically charged particles.
