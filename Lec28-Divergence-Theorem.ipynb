# Flux through a surface

Flux of $\vec F$ through surface S is $\iint_S \vec F \cdot \hat n\ dS$

Let S is graph of $z = f(x,y)$ and R is the shadow underneath in xy plane

$\hat n\ dS = \pm \langle -f_x, -f_y, 1 \rangle dx\ dy$. 

Plus minus for us to choose where is the convention direction

### Proof: $\hat n\ dS = \pm \langle -f_x, -f_y, 1 \rangle dx\ dy$

Let starts at a point $(x, y, f(x,y))$ and $\vec U = (x + \Delta x, y, f(x + \Delta x, y))$, $\vec V = (x, y + \Delta y, f(x, y + \Delta y))$

The cross product of $\vec U, \vec V$ will give us the area and the normal direction

$$\pm \vec U \times \vec V = \Delta S \cdot \hat n$$

$\vec U$ moves from $(x, y, f(x,y))$ to $(x + \Delta x, y, f(x + \Delta x, y) \approx f(x, y) + f_x \cdot \Delta x$

(Remember the change of f in x direction is equal to partial derivative f wrt x multiplies by change in x)

$\vec U \approx \langle \Delta x, 0, f_x \cdot \Delta x \rangle = \langle 1, 0, f_x \rangle \Delta x$

$\vec V \approx \langle 0, \Delta y, f_y \cdot \Delta y \rangle = \langle 0, 1, f_y \rangle \Delta y$

$$\hat n \cdot \Delta S = \vec U \times \vec V = \begin{vmatrix} \hat i & \hat j & \hat k \\ 1 & 0 & f_x \\ 0 & 1 & f_y \end{vmatrix} \Delta x \Delta y$$

$$= \langle -f_x, -f_y, 1 \rangle \Delta x \Delta y$$

positive when $\hat n$ upwards


### Example

$\vec F = z\hat k$ through the portion of paraboloid $z = x^2 + y^2$ above unit disk. Let choose normal upwards, then the flux should be positive

$$\vec F \cdot \hat n\ dS = \iint \langle 0,0,z \rangle \cdot \langle -2x, -2y, 1 \rangle dx dy$$

$f_x = 2x, f_y = 2y$ for $z = x^2 + y^2$

$$= \iint_S z\ dx\ dy = \iint_S (x^2 + y^2)\ dx\ dy$$

$$= \int_0^{2\pi} \int_0^1 r^2 \cdot r\ dr\ d\theta = \frac{\pi}{2}$$


## More generally

Let say z is very complicated and cannot be expressed in x and y. But we are able to parameterize it . That means we can express x, y, z in terms of any two parameter variables

Given parametric description of S

$$S: \begin{cases} x = x(u,v) \\
y = y(u,v) \\
z = z(u,v)
\end{cases}$$

$\langle x,y,z \rangle = \vec r = \vec r(u,v)$

We need to figure out the expression below

$$\hat n\ dS = ??\ du\ dv$$

One side is $\frac{\partial \vec r}{\partial u} \Delta u$, the other side is $\frac{\partial \vec r}{\partial v} \Delta v$

where $\frac{\partial \vec r}{\partial u} \Delta u = \langle \frac{\partial x}{\partial u} \Delta u, ... \rangle$

Therefore, $$\hat n\ dS = (\frac{\partial \vec r}{\partial u} \Delta u) \times (\frac{\partial \vec r}{\partial v} \Delta v) = (\frac{\partial \vec r}{\partial u} \times  \frac{\partial \vec r}{\partial v}) \Delta u \Delta v$$

$$\hat n\ dS = \pm (\frac{\partial \vec r}{\partial u} \times  \frac{\partial \vec r}{\partial v}) \Delta u \Delta v$$

plus minus depends on the direction we choose for convention



## Known Normal Vector

If we know a normal vector $\vec N$ (not necessarily unit) to the surface S

Example

1. A plane $ax + by + dz = d$, its normal vector $\vec N = \langle a,b,c \rangle$


2. S given by equation $g(x,y,z) = 0$, $\vec N = \nabla g$. Gradient is perpendicular to the level suface


## Normal Vector of Slanted Plane

Let say there is a slanted plane $\Delta S$, projected onto horizontal plane become $\Delta A$. The angle $\alpha$ is the angle that plane S rises from the plane A

Surface element: $\Delta A = \Delta S \cos\alpha$

Angle between vertical axis $\hat k$ and normal of the plane $\hat n$ is also $\alpha$

$$\cos\alpha = \frac{\vec N \cdot \hat k}{|\vec N| |\hat k|}$$

Hence

$$\Delta S = \frac{1}{\cos\alpha} \cdot \Delta A = \frac{|\vec N|}{\vec N \cdot \hat k} \Delta A$$

$$\hat n \cdot \Delta S = \frac{|\vec N| \cdot \hat n}{\vec N \cdot \hat k} \Delta A = \pm \frac{|\vec N|}{\vec N \cdot \hat k} \Delta A$$

plus minus depend on direction

$$\hat n \cdot dS = \pm \frac{|\vec N|}{\vec N \cdot \hat k}\ dx\ dy$$


### Example

The equation of surface S is $z - f(x,y) = 0$ or $g(x,y,z) = 0$

$\vec N = \nabla g = \langle -f_x, -f_y, 1 \rangle$

$$\frac{|\vec N|}{\vec N \cdot \hat k}\ dx\ dy = \langle -f_x, -f_y, 1 \rangle\ dx\ dy$$

The result is the same as the proof above

# Divergence Theorem

Also known as Gauss-Green theorem (3D analogue of Green for flux)

If S is a closed surface enclosing a region D, oriented with $\hat n$ outwards and $\vec F$ defined and differentiable everywhere in D, then $$\unicode{x222F}_S \vec F \cdot d\vec S = \iiint_D div \vec F\ dV$$

where $div(P\hat i + Q\hat j + R\hat k) = P_x + Q_y + R_z$

Last time, flux of $z\hat k$ through sphere radius a was $\frac43 \pi a^3$

$$\iint_S z\hat k \cdot \hat n\ dS = \iiint_D div(z\hat k)\ dV = \iiint_D 0+0+1\ dV = \iiint_D 1\ dV$$

which is the volume D

Physcially, the LHS of divergence theorem is the total amount of stuff that goes out of the region per unit time

The divergence measures how much the flow is expanding things. In a region of space, the total amount of water that flows out of it is the total amount of sources minus the sinks
