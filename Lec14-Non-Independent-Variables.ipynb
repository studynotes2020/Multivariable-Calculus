# Non Independent Variables

This lecture is about relations between variables, and how to handle functions that depend on several variables when they are related

Example: $f(P, V, T)$ where $PV = nRT$. What is the rate of change f, p and v wrt t or wrt to each others

Function of x, y, z, $f(x, y ,z)$ where the variables are related to each other by some equation $g(x, y, z) = c$. Same setup as last lecture, except we don't just concern on maxima and minima but to understand partial derivatives

If $g(x, y, z) = c$, then $z = z(x, y)$, what are $\frac{\partial z}{\partial x}, \frac{\partial z}{\partial y}$?


Example: $$x^2 + yz + z^3 = 8$$ at (2, 3, 1)

Take differential wrt x, y, z

$$2x \ dx + z \ dy + (y + 3z^2)\ dz = 0$$

Because g = 8, dg = 0. At (2, 3, 1)

$$4\ dx + dy + 6\ dz = 0$$

If we view $z = z(x, y)$

$$dz = -\frac{1}{6} (4\ dx + dy)$$

$$\frac{\partial z}{\partial x} = -\frac{2}{3}, \quad \frac{\partial z}{\partial y} = -\frac{1}{6}$$

To get the first expression, we set y constant, dy = 0. Second expression, x is constant, dx is zero

In general, $g(x, y, z) = c$, then 

$$dg = g_x \ dx + g_y \ dy + g_z \ dz = 0$$

Solve for dz:

$$dz = -\frac{g_x}{g_z} \ dx - \frac{g_y}{g_z} \ dy$$

So, for $\frac{\partial z}{\partial x}$, set y is constant, dy = 0

$$dz = -\frac{g_x}{g_z}\ dx \implies \frac{\partial z}{\partial x} = -\frac{g_x}{g_z}$$

Say $f(x, y) = x + y, \frac{\partial f}{\partial x} = 1$

Let change variables, $x = u, y = u + v$, then

$$f = 2u + v, \quad \frac{\partial f}{\partial u} = 2$$

So $x = u$ but $\frac{\partial f}{\partial x} \neq \frac{\partial f}{\partial u}$. How to resolve this contradiction?

It is because one is keeping y constant and vary x, another is keeping v constant and vary u

Need clearer notation:

$$\left(\frac{\partial f}{\partial x}\right)_y$$

means keep y constant

$$\left(\frac{\partial f}{\partial u}\right)_v$$

means keep v constant

$$\left(\frac{\partial f}{\partial x}\right)_y \neq \left(\frac{\partial f}{\partial x}\right)_v = \left(\frac{\partial f}{\partial u}\right)_v$$

### Example: Area of Triangle

A triangle with area

$$A(a, b, \theta) = \frac12 ab \sin \theta$$

Assume it's a right triangle with $a = b\cos \theta$ which is the constraint (relation between a, b, theta)

Rate of change of $A$ wrt $\theta$

1. Treat $a, b, \theta$ as independent, 

$$\frac{\partial A}{\partial \theta} = \left(\frac{\partial A}{\partial \theta}\right)_{a,b}$$

keep a, b fixed

$$\frac{\partial A}{\partial \theta} = \frac12 a b \cos \theta$$

2. Keep $a$ constant, $b$ will change $b = b(a, \theta) = \frac{a}{\cos \theta}$. So that we keep a right angle

$$\left(\frac{\partial A}{\partial \theta}\right)_{a}$$

3. Keep $b$ constant, $a = a(b, \theta)$ changes

$$\left(\frac{\partial A}{\partial \theta}\right)_{b}$$


### Compute $\left(\frac{\partial A}{\partial \theta}\right)_{a}$

Method 0: Solve for b, and substitute

$a = b\cos \theta \implies b = \frac{a}{\cos \theta} = a \sec \theta$

$$A = \frac12 ab\sin\theta = \frac12 \frac{a^2 \sin\theta}{\cos\theta} = \frac12 a^2 \tan\theta$$

$$\left(\frac{\partial A}{\partial \theta}\right)_{a} = \frac12 a^2 \sec^2\theta$$

Two systematic methods:

1. Differentials
    - Keep a fixed, da = 0
    - constraint $a = b\cos\theta$
    - $da = \cos\theta \ db - b\sin\theta \ d\theta \implies \cos\theta \ db = b\sin\theta \ d\theta \implies db = b\tan\theta \ d\theta$
    
Function, $A = \frac12 ab\sin\theta$, 

$$dA = \frac12 b\sin\theta \ da + \frac12 a \sin\theta \ db + \frac12 ab \cos\theta \ d\theta$$

This function tells us how dA varies with da, db, $d\theta$

We have the result of da and db, substitute in

$$dA = \frac12 a\sin\theta (b \tan\theta \ d\theta) + \frac12 ab \cos\theta \ d\theta = \frac12 ab (\sin\theta \tan\theta + \cos\theta) \ d\theta$$

Because $\sin\theta \tan\theta + \cos\theta = \sec\theta$

$$dA = \frac12 ab \sec\theta \ d\theta$$

$$\left(\frac{\partial A}{\partial \theta}\right)_{a} = \frac12 ab \sec\theta$$


Summary of method 1

- Write dA in terms of da, db, $d\theta$
- a is constant, then da = 0
- differentiate constraint, solve for db in terms of $d\theta$
- Plug into dA, get answer



2. Chain Rule

$\left(\frac{\partial }{\partial \theta}\right)_{a}$ in formula for $A$

$$\left(\frac{\partial A}{\partial \theta}\right)_{a} = A_{\theta} \left(\frac{\partial \theta}{\partial \theta}\right)_{a} + A_a \left(\frac{\partial a}{\partial \theta}\right)_{a} + A_b \left(\frac{\partial b}{\partial \theta}\right)_{a}$$

$\left(\frac{\partial \theta}{\partial \theta}\right)_{a} = 1, \quad \left(\frac{\partial a}{\partial \theta}\right)_{a} = 0, \quad \left(\frac{\partial b}{\partial \theta}\right)_{a}$ can be found by constraint
