# Recall Curl

$$curl\vec F = \begin{vmatrix}\hat i & \hat j & \hat k \\
\frac{\partial }{\partial x} & \frac{\partial }{\partial y} & \frac{\partial }{\partial z} \\
P & Q & R
\end{vmatrix} = \langle R_y-Q_z, P_z-R_x, Q_x-P_y \rangle$$

If imagine the vector fields measure the velocity in some fluids, then the curl measures how much the rotation in the vector fields

Direction corresponds to the axis of rotation, and the magnitude corresponds to twice the angular velocity

If taking constant velocity fields, that everything translates at the same speed, $\vec F = \langle a, b, c \rangle \implies curl \vec F = 0$

For an expanding velocity fields, $\vec F = \langle x, 0, 0 \rangle \implies curl \vec F = 0$. But $div\vec F = 1$. Divergence measures the stretching effect in the field

For a rotating velocity field, $\vec F = \langle -y, x, 0 \rangle \implies curl \vec F = 2\hat k$



# Stokes' Theorem

$$\oint_c \vec F \cdot d\vec r = \iint_S (\nabla \times \vec F) \cdot \hat n\ dS$$ 

$\nabla \times \vec F = curl\vec F$

C is closed surface. S is any surface bounded by C, and $\vec F$ is defined everywhere on S

Orientation: Need orientations of C and S to be compatible

If I walk along C in the positive direction, with S on my left, then $\hat n$ is pointing up

Or with right-hand rule: thumb along C positively, index finger tangent to S (towards interior of S), middle finger points parallel to $\hat n$

Example: A dome S covers a curve C, rotate from left to right. Then thumb points from left to right, index points to front, middle points up which is the direction of the normal vector

Example: An ice cream cone. Thumb goes from left to right, index goes down, normal vector points up and in 

Example: A cylinder. Two boundary curves, top C and bottom C'. How to choose the orietantion of the curves so that the normal vector $\hat n$ sticks out? C' (bottom curve) should go counterclockwise so that the cylinder is on your left. Check with right hand rule, thumb points counterclockwise, index finger points up and middle finger points out of cylinder

Same argument, top curve C should go clockwise


### Comparing Stokes with Green:

S is a portion of xy plane, bounded by a curve C (counterclockwise)

$$\oint_c \vec F \cdot d\vec r = \oint_c P\ dx + Q\ dy$$

$\vec F = \langle P,Q,R \rangle$ and with no component z -> z = 0, dz = 0

Stokes said

$$= \iint_S (\nabla \times \vec F)\cdot \hat n\ dS$$

From the right hand rule, the normal vector points up. $\hat n = \hat k$

$$= \iint_S (\nabla \times \vec F)\cdot \hat k\ dS$$

$(\nabla \times \vec F)\cdot \hat k$ is the z component curl $\vec F$

$$= \iint_S (Q_x - P_y)\ dx\ dy$$

$$\oint_c P\ dx + Q\ dy = \iint_S (Q_x - P_y)\ dx\ dy$$

is exactly the statement of Green's theorem

Green's theorem is a special case of Stokes in the xy plane!!

We choose right hand coordinate system and direction of cross product so that the direction is consistent



## Why Stokes is true?

- know it for C, S in xy plane

- also for C, S in any plane (using that work, flux, curl make sense independently of the coordinate system)

- given any S, decompose it into tiny, almost flat pieces. Sum of work around each piece = work along C. Sum of flux through each piece = flux through S

In short, from Stokes' theorem for flat surfaces, we can get it for any surface


### Example

Find work of $\vec F = z\hat i + x\hat j + y\hat k$ around unit circle on xy plane (counterclockwise)

Directly: $$\oint z\ dx + x\ dy + y\ dz$$

Parameterize: $x = \cos t, y = \sin t, z = 0$

$$\int 0 + \int_0^{2\pi} \cos t . (\cos t\ dt) + \int 0 = \int_0^{2\pi} \cos^2t\ dt = \pi$$
