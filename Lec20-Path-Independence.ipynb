# Last Lecture

We studied line integrals for work

$$\int_c \vec F \cdot d\vec r = \int_c \vec F \cdot \hat T \ ds = \int_c M\ dx + N\ dy$$

where $\vec F = \langle M, N \rangle$

Example: $\vec F = \langle y, x \rangle$, $c = c_1 + c_2 + c_3$ is a trajectory encloses a sector of unit disk $0 \leq \theta \leq \frac{\pi}{4}$ (Picture shown in the lecture is that c1 move from left to right, c2 move curvely upwards and c3 moves back to the origin)

Need to setup $\int_{c_i} y\ dx + x\ dy$

1. At x-axis: from (0,0) to (1,0). y = 0, dy = 0. 

$$\int_{c_1}y\ dx + x\ dy = \int_{c_1} 0\ dx + 0 = 0$$

Geometrically, on x-axis, $\vec F$ is parallel to $\hat j$, $\vec F \cdot \hat T = 0$

2. $c_2$ is a portion of unit circle. $x = \cos\theta$, $y = \sin\theta$, $0 \leq \theta \leq \frac{\pi}{4}$. Hence $dx = -\sin\theta\ d\theta$, $dy = \cos\theta\ d\theta$

$$\int_{c_2}y\ dx + x\ dy = \int_0^{\pi/4} \sin\theta (-\sin\theta)\ d\theta + \cos\theta (\cos\theta)\ d\theta$$

$$ = \int_0^{\pi/4} \cos^2\theta - \sin^2\theta\ d\theta = [\frac12 \sin(2\theta)]_0^{\pi/4} = \frac12$$

3. $c_3$ starts from $(\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}})$ goes back to (0,0). We want to find $\int_{c_3}y\ dx + x\ dy$

Could parametrize the curve: $x = \frac{1}{\sqrt{2}} - \frac{1}{\sqrt{2}}t$, $y = \frac{1}{\sqrt{2}} - \frac{1}{\sqrt{2}}t $, $ 0 \leq t \leq 1$. This is fine but unnecessarily complicated.

We can instead see the motion as going opposite, which is from (0,0) to $(\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}})$

Better: $x = t, y = t$, $0 \leq t \leq \frac{1}{\sqrt{2}}$ gives us $-c_3$ and $\int_{-c_3} = -\int_{c_3}$

$$\int_{-c_3}y\ dx + x\ dy = \int_0^{1/\sqrt{2}} t\ dt + t\ dt = \int_0^{1/\sqrt{2}} 2t\ dt$$

$$= [t^2]_0^{1/\sqrt{2}} = \frac12$$

Hence the opposite direction

$$\int_{c_3}y\ dx + x\ dy = -\frac12$$


4. Total work = $\int_c = \int_{c_1} + \int_{c_2} + \int_{c_3} = 0 + 1/2 - 1/2 = 0$

## How to avoid computing line integrals

Special case: say $\vec F = \nabla f$ is gradient field. $f(x, y)$ is called potential. Then we can simplify evaluation of $\int_c \vec F \cdot d\vec r$

Similar to Fundamental of Calculus which states that integrate a derivative gets back the function. 

Fundamental theorem of Calculus for line integrals of the gradient of a function, gets back the function. 

$$\int_c \nabla f \cdot d\vec r = f(p_1) - f(p_0)$$

where trajectory $c$ moves from $p_0$ to $p_1$. Or write it in another way

$$\int_c f_x\ dx + f_y\ dy = \int_c df = f(p_1) - f(p_0)$$

### Proof 

$$\int_c \nabla f \cdot d\vec r = \int_c f_x\ dx + f_y\ dy $$

We want to parametrize $c$ into $x = x(t), y = y(t)$, then $dx = x'(t)\ dt, dy = y'(t)\ dt$

$$\int_c \nabla f \cdot d\vec r = \int_c (f_x \frac{dx}{dt}+ f_y \frac{dy}{dt}) dt = \int_{t_0}^{t_1} \frac{df}{dt}\ dt$$

Last step is by Chain rule. 

$$\int_{t_0}^{t_1} \frac{df}{dt}\ dt = \left[f(x(t), y(t)) \right]_{t_0}^{t_1} = f(p_1) - f(p_0)$$


### Example

$\vec F = \langle y, x \rangle = \nabla f \implies f(x,y) = xy$ because derivative f wrt x is y and f' wrt y is x

Then we can compute line integrals by finding the values of f at the endpoint. The motion starts from (0,0), $f(0,0) = 0$, then $c_1$ moves along x-axis, $f(1, 0) = 0$. So, there is no work, the potential doesn't change. 

Then for $c_2$, the potential increases to $\frac12$. 

$$\int_{c_2}\vec F \cdot d\vec r = f(\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}) - f(1, 0) = \frac12$$

For $c_3$, $f(p_0) - f(p_2) = 0 - \frac12$. The work goes back down form 1/2 to zero, the work is negative 1/2


A lot of forces are gradients of potentials, namely gravitational force, electric force. But not every vector field is a gradients. For example, magnetic fields are not vector fields

WARNING: Everything today only applies if $\vec F$ is a gradient field! Not true otherwise!

### Consequences of fundamental theorem

If $\vec F$ is a gradient field then

- Path independence. For example, the trajectory moves from p0 to p1, via c1 or c2, the line integrals of c1 is equal to line integrals of c2. This means it doensn't matter the path, if c1, c2 have the same start and end points

$$\int_{c_1} \nabla f \cdot d\vec r = \int_{c_2} \nabla f \cdot d\vec r$$

- $\vec F = \nabla f$ is conservative. If you take a close trajectory $c$, then the work done along c, $\int_{c}\vec F \cdot d\vec r = 0$. Non-conservative means there exsits some enclosed curves such that work done is not zero

If you have force field that is conservative, you cannot build some perpetual motion. You can't build something that is keep going, powered by that force because the force is not providing any energy. After finishing one loop, nothing is happened from the point of view of the energy provided by the force.

If the force field is non-conservative, then we can try to find a loop where the work would be positive. The motion will just keep running by itself. For example magnetic fields and transformers or power adapters, you extract energy from the magnetic field. But then you need power supply to maintain the magnetic fields. 


$$\int_{c=closed} \nabla f \cdot d\vec r = f(end) - f(start) = 0$$

where end and start are the same points


Remark: $\vec F = \langle -y, x \rangle$ and take c is unit circle. On unit circle F is parallel to tangent. The length F on unit circle is 1 and unit tangent has length 1

$$\int_c \vec F \cdot \hat T\ ds = \int_c |\vec F|\ ds = \int_0^{2\pi} 1\ ds = 2\pi$$

$2\pi$ is not 0, so this vector field is not conservative, which implies it is not the gradient of anything

It is not path independence either. Let say we go from (1,0) to (-1,0). If we go anticlockwise path, we get $\pi$. If we go clockwise path, we get $-\pi$, which is not the same

Conservative and path independence are equivalence.



## Physics

If force field $\vec F$ is gradient of a potential, $\vec F = \nabla f$, then work of $\vec F$ is the change in value of potential. E.g. gravitational field, electrical field vs gravitational and electrical potential.

Conservativeness means no energy can be extracted from the field "for free"

Total energy is conserved

### EQUIVALENT PROPERTIES

- $\vec F$ is conservative. $\int_{c} \vec F \cdot d\vec r = 0$ along all closed curve c

- $\int_{c} \vec F \cdot d\vec r$ is path independence

Non convervative means if c1 and c2 moves to a point with different path, then c1 - c2 is not zero

- $\vec F$ is a gradient field, $\vec F = \langle f_x, f_y \rangle$

$\impliedby$ is fundamental theorem

$\implies$ will be how we find the potential. We find potential by just computing how it changes from start to end, without concerning the path of change

- $M\ dx + N\ dy$ is an exact differential = df
