# Spherical Coordinates

Use distance from the origin then use angles to determine which direction you are heading

Distance from the origin $\rho$, 0 to infinity

Angle downwards from z-axis $\phi$, 0 to $\pi$

Angle counterclockwise from x-axis $\theta$, 0 to $2\pi$

On a sphere $\rho = a$

$\phi$ tells us about latitude (how far from the north pole)

$\theta$ tells us about longtitude (how far in east or west)

The idea of spherical coordinate is to do polar coordinate in rz plane

In a rz plane, $\phi$ is the angle between $\rho$ and z-axis

$z = \rho \cos\phi$

$r = \rho\sin\phi$ 

We can use these two relationships to switch between polar and spherical coordinate

To switch between xyz space

$x = r\cos\theta = \rho\sin\phi\cos\theta$

$y = r\sin\theta = \rho\sin\phi \sin\theta$

$z = \rho\cos\phi$

$\rho = \sqrt{r^2 + z^2} = \sqrt{x^2 + y^2 + z^2}$


- $\rho = a$. This is a sphere of radius a centered at origin


- $\phi = \frac{\pi}{4}$ is a cone. Because the equation does not involve theta, it is all the same when rotated along vertical z axis. If additionally give $z = r$, then this cone centered around r


- $\phi = \frac{\pi}{2}$ is xy plane

## Triple Integral in Spherical Coordinates

$dV = ???\ d\rho\ d\phi\ d\theta$

$\Delta V = ???\ \Delta\rho \Delta\phi \Delta\theta$

What is the surface area on sphere of radius a?

Piece of circle of radius $r = \rho\sin\phi = a\sin\phi $

East-west length or changes in theta direction is $a\sin\phi \Delta \theta$

North-south length or changes in phi direction is $a \Delta \phi$

To recap, we have a small piece of area where width is $a\sin\phi \Delta \theta$ and height is $a \Delta \phi$

$\Delta S = (a\sin\phi \Delta \theta) (a \Delta \phi) = a^2 \sin\phi \Delta \phi \Delta\theta$

$dS = a^2 \sin\phi\ d\phi\ d\theta$

Extending to volume

$\Delta V = \Delta \rho \cdot \Delta S = \Delta\rho \cdot \rho^2 \sin\phi  \Delta\phi \Delta\theta$

$dV = \rho^2 \sin\phi\ d\rho\ d\phi\ d\theta$

When doing triple integral for dx dy dz, it will become equation above for spherical coordinate


### Example

Volume of portion of unit sphere above $z = \frac{1}{\sqrt{2}}$

$$\iiint 1.dV = \iiint \rho^2 \sin\phi\ d\rho\ d\phi\ d\theta$$

Plane at $z = \frac{1}{\sqrt{2}} \implies \rho\cos\phi = \frac{1}{\sqrt{2}} \implies \rho = \frac{1}{\sqrt{2} \cos\phi}$. So $\rho$ changes from this point to the top where $\rho = 1$

$$\int_0^{2\pi} \int_0^{\frac{\pi}{4}} \int_{\frac{\sec\phi}{\sqrt{2}}}^{1} \rho^2 \sin\phi\ d\rho\ d\phi\ d\theta$$

Imagine of rz plane: $\rho\cos\phi = \frac{1}{\sqrt{2}} \implies \cos\phi = \frac{1}{\sqrt{2}} \implies \phi = \frac{\pi}{4}$. So $\phi$ changes from the top at 0 to $\pi / 4$

We go all around the sphere so theta from 0 to 2pi

Evaluation: $$\int_0^{2\pi} \int_0^{\frac{\pi}{4}} \int_{\frac{\sec\phi}{\sqrt{2}}}^{1} \rho^2 \sin\phi\ d\rho\ d\phi\ d\theta = \frac{2\pi}{3} - \frac{5\pi}{6\sqrt{2}}$$


### Applications: Gravitational attraction

Gravitational force exerted by $\Delta M$ at (x,y,z) on a mass m at origin at a distance $\rho$

Magnitude of the force

$$|\vec F| = G \cdot \frac{\Delta M \cdot m}{\rho^2}$$

unit vector $\vec F = \frac{\langle x,y,z \rangle}{\rho}$. vector = magnitude x unit vector

$$\vec F = G \cdot \frac{\Delta M \cdot m}{\rho^3} \langle x,y,z \rangle$$

Integrating: Mass = density x volume

$$\Delta M = \delta \Delta V$$

$$\vec F = \iiint G \cdot \frac{\delta \Delta V \cdot m}{\rho^3} \langle x,y,z \rangle $$

Integrate for each x, y, z components

Setup: place a solid so z-axis is an axis of symmetry, then $\vec F = \langle 0,0,F \rangle$

z component: $$Gm \iiint \frac{z}{\rho^3} \delta\ dV$$

Use spherical coordinates

$$F_z = Gm \iiint \frac{\rho\cos\phi}{\rho^3} \delta \cdot \rho^2 \sin\phi\ d\rho\ d\phi\ d\theta = Gm \iiint \cos\phi \cdot \delta \cdot \sin\phi\ d\rho\ d\phi\ d\theta$$


Example: Newton theorem of gravitation attraction of a spherical planet with uniform density is equal to that of a point mass with the same total mass at its center
