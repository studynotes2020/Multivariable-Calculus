# Vector Field in Space

$\vec F = \langle P, Q, R \rangle$

P, Q, R are functions of x, y, z 

1. Example: Force fields (gravitational attraction of a solid mass at (0,0,0) on a mass m at (x,y,z))

$\vec F$ directed towards origin, magnitude will be order of $\frac{c}{\rho^2}$. $\rho$ is the distance from the origin

$$\vec F = -c \frac{\langle x, y, z \rangle}{\rho^3}$$

positive go outwards from origin, negative points towards origin 

Electric fields, magnetic fields, ...

2. Another example is velocity fields $\vec v$. Fluids flow. Wind in not the same at all altitudes


3. Gradient fields. $u = u(x, y, z), \nabla u = \langle u_x, u_y, u_z \rangle$

Above are not mutually exclusive


## Flux

Recall, in 2D, flux is line integrals $\int_c \vec F \cdot \hat n\ ds$

In 3D, flux will be measured through a surface. Surface integrals

We will be summing over all the pieces of a surface. In line integrals in the lane, we have two variables and reduce to one. For surface integral, we have three variables and reduce to two

$\vec F$ vector field, S surface in space. $\hat n$ is unit normal vector to S. There are 2 choices for $\hat n$. Choose a side for the surface (orientation)

$$Flux = \iint_S \vec F \cdot \hat n\ dS$$

dS is surface area element. Use dS rather than dA because he want to think of dA is the area of the coordinate plane. But conceptually they are similar

Cut the surface into many small pieces where each piece $\Delta s$, take the dot product of vector field with unit normal vector. Then sum all the dot products

If surface tangent to the flux, then flux is zero

Notation: $d\vec S = \hat n\ dS$ can be written directly as (dy, -dx), easier to compute


### Example 1

1. Flux of $\vec F = \langle x, y, z \rangle$ through sphere of radius a centered at origin

F is pointing away from the origin, parallel to normal vector

$$\iint_S \vec F \cdot d\vec S = \iint_S \vec F \cdot \hat n\ dS$$

$\hat n = \frac{1}{a}\langle x, y, z \rangle$ Divide by length a, to get unit length (length is $\sqrt{x^2 + y^2 + z^2} = a$)

$\vec F \cdot \hat n = |\vec F|.1 = a$ since $\vec F$ parallel to $\hat n$

Or see this geometrically,

$$\vec F \cdot \hat n = \langle x, y, z \rangle \cdot \frac{1}{a}\langle x, y, z \rangle = \frac{a^2}{a} = a$$

Until now, everything is done on the surface. Nothing on inside and outside of the sphere

$$\vec F \cdot \hat n = \iint_S a\ dS = a \iint_S  dS = 4\pi a^3$$


### Example 2

Same sphere, with vector field $\vec H = z \hat k$

We still have $\hat n =  \langle x, y, z \rangle / a$

$$\vec H \cdot \hat n = \langle 0,0,z \rangle \cdot \langle x, y, z \rangle / a = \frac{z^2}{a}$$

$$\iint_S \vec H \cdot \hat n\ dS = \iint_S \frac{z^2}{a}\ dS$$

We need to figure out what dS is

$\Delta S = (a\sin\phi \Delta\theta).(a\Delta\phi) \implies dS = a^2\sin\phi\ d\phi\ d\theta$

Here we are finding a parametric equation for the sphere using two variables phi and theta, which happen to be part of the spherical coordinate system

$z = a\cos\phi$ because phi is the angle bounds to z-axis and the distance is a

$$\iint_S \frac{z^2}{a}\ dS = \int_0^{2\pi} \int_0^{\pi} \frac{a^2\cos^2\phi}{a} \cdot a^2\sin\phi\ d\phi\ d\theta$$

We go all the way from the north pole to south pole so phi from 0 to pi and then theta goes from 0 to 2pi

$$= 2\pi a^3 \cdot [-\frac13 \cos^3\phi]_0^{\pi} = \frac43 \pi a^3$$


## Conclusion

Use geometry or setup $\iint_S \vec F \cdot \hat n\ dS$

0. Easy one. S is horizontal plane, $z = a$. $\hat n = \pm \hat k, dS = dx\ dy$

For vertical plane parallel to yz plane, then $x = a$, $\hat n = \pm \hat i, dS = dy\ dz$


1. Sphere of radius a from origin. $\hat n = \pm \langle x,y,z \rangle / a, dS = a^2\sin\phi\ d\phi\ d\theta$

Advice: Do the dot product first, if see things are symmetry, turns to phi and theta


2. Cylinder of radius a centered at the z-axis. $\hat n = \pm \frac{1}{a} \langle x,y,0 \rangle, dS = a \ dz\ d\theta $


3. Graph $z = f(x,y)$ $\hat n\ dS =  \pm \langle -f_x, -f_y, 1 \rangle dx\ dy$

$\pm \langle -f_x, -f_y, 1 \rangle$ is not $\hat n$

$dx\ dy$ is not dS

Each of the is more complicated and the combination somehow simplifies nicely

To set up bounds on $\iint \ dx\ dy$ look at shadow of S in xy plane. Do this when z is eliminated


## Geomtric interpretation

If $\vec F$ is a velocity field, then the flux is the amount of matter through S per unit time

What is the stuff flows through in a second?

It is the stuff lives in a small box whose base is that small piece of surface and 4 other sides are given by the vector field. The volume is base times height and the height is $\vec F \cdot n$
