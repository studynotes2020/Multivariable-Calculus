# Recall

Chain rule

$w = w(x, y, z)$

$x = x(t), \quad y = y(t), \quad z = z(t)$

$$\frac{dw}{dt} = w_x \frac{dx}{dt} + w_y \frac{dy}{dt} + w_z \frac{dz}{dt}$$

The changed in w is caused by changes in x, y, z

$$\frac{dw}{dt} = \nabla w \cdot \frac{d \vec r}{dt}$$

$$\nabla w = \langle w_x, w_y, w_z \rangle$$

$\nabla w$ is the gradient of w at some point (x, y, z)

$$\frac{d \vec r}{dt} = \langle \frac{dx}{dt}, \frac{dy}{dt}, \frac{dz}{dt} \rangle$$

## Theorem:  $\nabla w \bot$ level surface (w = constant)

Example 1:

$w = a_1x + a_2y + a_3z$

$$\nabla w =  \langle a_1, a_2, a_3 \rangle = \langle \frac{\partial w}{\partial x}, \frac{\partial w}{\partial y}, \frac{\partial w}{\partial z} \rangle$$

level surface $a_1x + a_2y + a_3z = c$, plane with normal $\langle a_1, a_2, a_3 \rangle$

Example 2:

$w = x^2 + y^2$. Level curve $w = c$ is a circle $x^2 + y^2 = c$

$\nabla w = \langle 2x, 2y \rangle $ which is a line perpendicular to the circle. It points towards higher values of w

### Proof

Take a curve $\vec r = \vec r (t)$ that stays on the level $w = c$

$w(x, y, z) = c$

velocity vector $\vec v = \frac{d \vec r}{dt}$ is tangent to the level $w = c$

By chain rule:

$$\frac{dw}{dt} = \nabla w \cdot \frac{d \vec r}{dt} = \nabla w \cdot \vec v = 0$$

because $w(t) = c$ constant

The dot product is zero tells us that two vectors are perpendicular, so $\nabla w \bot \vec v$

This is true for any motion on $w = c$. $\vec v$ can be any vector tangent to $w = c$

Given any vector $\vec v$ on tangent plane to level curve $w = c$,  $\nabla w \bot \vec v$

So  $\nabla w \bot$ tangent plane to the level

For 2D, gradient is perpendicular to the tangent line

Example: Find the tangent plane to surface $x^2 + y^2 - z^2 = 4$ at (2, 1, 1) 

Level set $w = 4$, where $w = x^2 + y^2 - z^2$

Gradient (normal vector) $\nabla w = \langle 2x, 2y, -2z \rangle = \langle 4, 2, -2 \rangle$

So equation is $4x + 2y - 2z = 8$. So given a function (surface) and a point, we know how to find the tangent plane to the surface at that point


Another way:

$dw = 2x \ dx + 2y \ dy - 2z \ dz$, at (2,1,1)

$dw = 4 \ dx + 2 \ dy - 2 \ dz$

$\Delta w \approx 4 \Delta x + 2 \Delta y - 2 \Delta z$

So, the level corresponds to $\Delta w = 0$ and its tangent plane corresponds to $4 \Delta x + 2 \Delta y - 2 \Delta z = 0$

$4 (x - 2) + 2 (y - 1) - 2 (z - 1) = 0 \implies 4x + 2y - 2z = 8$

which is the same answer we get from the first method


## Directional Derivatives

$w = w(x, y)$, we know how to compute $\frac{\partial w}{\partial x}, \frac{\partial w}{\partial y}$ which measures how w changes if we move in the direction of x or in the direction of y

What about other directions? Is there a derivative in every direction? Yes, it is called directional derivatives

What if we move in the directioin of $\hat u$ unit vector?

Let look at straight line trajectory $\vec r(s), \frac{d \vec r}{ds} = \hat u$. 

$s$ is an arc length distance along line. 

What is $\frac{dw}{ds}$?

If $\hat u = \langle a, b \rangle$

$$\begin{cases}
x(s) = x_0 + as \\
y(s) = y_0 + bs
\end{cases}$$

plug into w and compute $\frac{dw}{ds}$

Definition: directional derivative in direction of $\hat u$

$$\frac{dw}{ds}_{| \hat u}$$

which is the slope of slice of graph by a vertical plane parallel to $\hat u$

Chain rule implies $$\frac{dw}{ds} = \nabla w \cdot \frac{d\vec r}{ds} = \nabla w \cdot \hat u$$

$$\frac{dw}{ds}_{| \hat u} = \nabla w \cdot \hat u$$

$\nabla w $ is the component of $\nabla w$ in direction of $\hat u$

Example: $$\frac{dw}{ds}_{| \hat i} = \nabla w \cdot \hat i = \frac{\partial w}{\partial x}$$


## Geometry

$$\frac{dw}{ds}_{| \hat u} = \nabla w \cdot \hat u = |\nabla w|.1.\cos\theta$$

the changes is maximal when $\cos\theta = 1 \implies \theta = 0 \implies \hat u = dir(\nabla w)$

The gradient is the direction where the function increases the most quickly at that point

So, direction of $\nabla w$ is the direction of fastest increase of w

$$|\nabla w| = \frac{dw}{ds}_{| dir(\nabla w)}$$

The minimal of $\frac{dw}{ds}_{| \hat u}$ is where $\cos\theta = 1 \implies \theta = 180^o$. $\hat u$ is in direction of $(-\nabla w)$

In which direction $\frac{dw}{ds}_{| \hat u} = 0$? When $\cos\theta = 0 \implies \theta = 90^o \implies \hat u \bot \nabla w \iff \hat u$ tangent to the level
