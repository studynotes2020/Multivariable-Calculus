# Triple Integrals

$$\iiint_R f\ dV$$

dV is volume element = dx dy dz

Example 1:

region between paraboloids $z = x^2 + y^2$ and $z = 4 - x^2 - y^2$

volume $\iiint 1\ dV$

It is easier to start with dz because how the question given. Then dx or dy first doesn't matter in this problem bcs they are symmetric

$$\int_{-\sqrt{2}}^{\sqrt{2}} \int_{-\sqrt{2-x^2}}^{\sqrt{2-x^2}} \int_{x^2 + y^2}^{4 - x^2 - y^2} \ dz\ dy\ dx$$

Find shadow in xy plane wherever z bottom < z top

$$x^2 + y^2 < 4 - x^2 - y^2$$

$$x^2 + y^2 < 2$$

which is disk of radius $\sqrt{2}$. Set x constant and y varies and we get y changes from $-\sqrt{2-x^2}$ to $\sqrt{2-x^2}$

Lastly, x changes from $-\sqrt{2}$ to $\sqrt{2}$ (Need to review how to setup double integral)

Inner integral: $$\int_{x^2 + y^2}^{4 - x^2 - y^2} \ dz = [z]_{x^2 + y^2}^{4 - x^2 - y^2} = 4 - 2x^2 - 2y^2$$

$$\int_{-\sqrt{2}}^{\sqrt{2}} \int_{-\sqrt{2-x^2}}^{\sqrt{2-x^2}}  (4 - 2x^2 - 2y^2) \ dy\ dx$$

Then switch to polar coordinate


## Polar Coordinate

Better use polar coordinate instead of xy

$$\int_0^{2\pi} \int_0^{\sqrt{2}}\int_{r^2}^{4-r^2} dz\ r\ dr\ d\theta$$

This is easier to evaluate



## Cylindrical coordinates

Instead of x, y, z, there will be $(r, \theta, z)$ 

Conversion between them

$x = r\cos\theta, y = r\sin\theta, z = z$

Note given $r = a$, we will get cylinder of radius a centered on z-axis

Volume element: $dx\ dy\ dz = r\ dr\ d\theta\ dz$

$\Delta v = \Delta A \cdot \Delta z$

$dV = dA \cdot dz$



## Applications

- Use to find Mass: 

Density  $$\delta = \frac{\Delta m}{\Delta V}$$

$$dm = \delta \cdot dV$$

$$Mass = \iiint_R \delta dV$$

- Average value of $f(x,y,z)$ in R:

$$\bar f = \frac{1}{Vol(R)} \iiint_R f \ dV$$

or with density (weighted average)

$$\frac{1}{Mass(R)} \iiint_R f \cdot \delta \ dV$$

- Center of mass: $(\bar x, \bar y, \bar z)$

$$\bar x = \frac{1}{Mass} \iiint x \cdot \delta \ dV$$

In example: $\bar x = \bar y = 0$ by symmetry!

- Moment of inertia of an axis

$$\iiint_R (\text{distance to axis})^2\ \delta\ dV$$

$$I_z = \iiint_R r^2 \delta \ dV = \iiint_R (x^2 + y^2) \delta \ dV$$

$$I_x = \iiint_R (y^2 + z^2 ) \delta \ dV$$

$$I_y = \iiint_R (x^2 + z^2 ) \delta \ dV$$



### Example 2

$I_z$ of a solid cone between $z = ar, z = b$. Density is 1

$$I_z = \iiint r^2 \cdot r\ dr\ d\theta\ dz$$

dz at last means we slice across z

Equation of cone: $z = ar \iff r = \frac{z}{a}$

$$I_z = \int_0^b \int_0^{2\pi} \int_0^{z/a} r^2 \cdot r\ dr\ d\theta\ dz = \frac{\pi b^5}{10a^4}$$

For the bound of dz, the first slice is 0, last slice is b

$$x = \pm \sqrt{2y - 2y^2}$$
