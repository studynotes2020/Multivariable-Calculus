# Green's Theorem

If C is a closed curve, enclosing a region R, counterclockwise, $\vec F$ vector field defined and differentiable in R, then 

$$\oint_c \vec F \cdot d\vec r = \iint_R curl(\vec F)\ dA$$

$$\oint_c M\ dx + N\ dy = \iint_R (N_x - M_y)\ dA$$

LHS is about the curve, RHS is about the region enclosed by the curve

Warning: only for closed curves

Example: Let C = circle of radius 1 centered at (2,0), counterclockwise

$$\oint_c ye^{-x}\ dx + (\frac12 x^2 - e^{-x})\ dy$$

Directly: $x = 2 + \cos\theta, y = \sin\theta, dx = -\sin\theta \ d\theta, dy = \cos\theta \ d\theta$

Substitute all these in and integreate, this is difficult

Using Green's theorem: Compute 

$$\iint_R curl(\vec F)\ dA = \iint_R (N_x - M_y)\ dA  = \iint_R (x + e^{-x}) - e^{-x}\ dA$$

R is the region enclosed by the curve, $N = \frac12 x^2 - e^{-x}, M = ye^{-x}$

$$= \iint_R x\ dA = Area(R) \cdot \bar x = 2\pi$$

Because center of mass $\bar x = \frac{1}{Area} \iint x\ dA \implies \iint x\ dA = Area \cdot \bar x$

Center of mass at 2 and $A = \pi r^2$ for unit circle is $\pi$

### Special Case

If curl $\vec F = 0$, then is $\vec F$ conservative?

Green's: $$\oint_c \vec F \cdot d\vec r = \iint_R curl(\vec F)\ dA = 0$$

if curl $\vec F =  0$. This proves if curl $\vec F = 0$ everywhere in R, then $\oint_c \vec F \cdot d\vec r = 0$

Consequence: if $\vec F$ is defined everywhere in the plane and curve $\vec F = 0$, then $\vec F$ is conservative

Proof: $$\oint_c \vec F \cdot d\vec r = \iint_R curl(\vec F)\ dA = \iint_R 0\ dA = 0$$

Cannot apply Green's theorem to the vector field on PSet 8 Problem 2 when C encloses the origin! Because at origin, vector field is not defined


## Proof of Green's Theorem

$$\oint_c M\ dx + N\ dy = \iint_R (N_x - M_y)\ dA$$

First observation: we will prove $\oint_c M\ dx = \iint_R -M_y\ dA$. This is special case when $N = 0$

After we prove this, we can do similar procedure to $M = 0$. A similar argument will show 

$$\oint_c N\ dy = \iint_R N_x\ dA$$

Then we can combine the result to get Green's Theorem.

Second observation: we can decompose R into simpler regions! 

Let say there is a curve C, then we decompose it into curve c1 with region R1 and curve c2 with region R2

If we prove $\oint_{c_1} M\ dx = \iint_{R_1} -M_y\ dA$ and $\oint_{c_2} M\ dx = \iint_{R_2} -M_y\ dA$

There is a line in the middle which is passed twice, once from c1 and once from c2. Since they are in opposite direction, when we sum it, the effect will be cancelled

then $\oint_c M\ dx = \oint_{c_1} + \oint_{c_2} = \iint_{R_1} + \iint_{R_2} = \iint_{R} -M_y\ dA$

The first step valid because we go twice along boundary between R1 and R2 with opposite orientation

Cut R into "vertically simple" regions, $a < x < b$, $f_1(x) < y < f_2(x)$

### Main Step: prove $\oint_c M\ dx = \iint_R -M_y\ dA$ 

R is vertically simple and C is boundary of R counterclockwise

A picture drawn in the lecture. The closed curve C = c1 + c2 + c3 + c4. Top of the region is bounded by c3 or $f_2(x)$, bottom of the region is bounded by c1 or $f_1(x)$. Two vertical lines by the sides are c2 (going up) and c4 (going down)

$$\int_{c_1}M\ dx = \int_a^b M(x, f_1(x))\ dx$$

We don't know what M is, so we stop here

$$\int_{c_2}M\ dx = \int_a^b M(x, y)\ dx = 0$$

For line c2, x is b and doesn't change, so dx = 0. Similarly, for c4 

$$\int_{c_4}M\ dx = 0$$

For c3, $y = f_2(x)$ and x changes from b to a

$$\int_{c_3}M\ dx = \int_b^a M(x, f_2(x))\ dx = -\int_a^b M(x, f_2(x))\ dx$$

Summing the result

$$\oint_c M\ dx = \int_a^b M(x, f_1(x))\ dx - \int_a^b M(x, f_2(x))\ dx$$

This is the result of LHS. Then we moves to RHS

$$\iint_R -M_y\ dA = - \int_a^b \int_{f_1(x)}^{f_2(x)} \frac{\partial M}{\partial y}\ dy\ dx$$

Inner integral

$$\int_{f_1(x)}^{f_2(x)} \frac{\partial M}{\partial y}\ dy = M(x, f_2(x)) - M(x, f_1(x))$$

Hence 

$$\iint_R -M_y\ dA = -\int_a^b [M(x, f_2(x)) - M(x, f_1(x))] \ dx$$

which is equal to LHS. Note that this is true without knowing what is M

After we get the prove for vertically simple, we can do the same thing with y component. Sum the result, we have the full Green theorem

There is a device call planimeter that compute line integral to determine the area of an arbitrary two-dimensional shape
